```python
## cargar los paquetes de python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## cargar los datos
datos=pd.read_csv('milk.txt', sep='\s+')
datos
```

```python
### estadisticas descriptivas
datos.describe()
```

```python
## promedio
np.mean(datos['density'])
```

```python
## varianza
np.var(datos['density'])
```

```python
## desviación estándar
np.std(datos['density'])
```

```python
## quantile o percentiles
np.quantile(datos['density'], q=0.5)
## muestra el dato donde se acumula cierto porcentaje q
## donde  0<= q <=1
```


## Histogramas de la variable protein
# Boxplot de la variable protein
### Gráficos por paneles seaborn

Ejemplo de una ecuación:
$$\bar{x} = \sum^{n}_{i=1} x_{i} /n$$





```python
### grafico histograma

plt.hist(datos['protein'], color='gold')
plt.title('Histograma de la variable protein')
plt.ylabel('ni')
plt.xlabel('protein')
plt.text(35,15, 'hola')
```

```python
## Diagrama de cajas para la variable protein
plt.boxplot(datos['protein'])
plt.title('Diagrama de cajas de la variable protein')
plt.ylabel('protein')
plt.text(1.2, 33, 'UV')
```

```python
## Diagrama de cajas para la variable density
plt.boxplot(datos['density'])
plt.title('Diagrama de cajas de la variable density')
plt.ylabel('density')
plt.text(1.2, 1.030, 'UV')
```

```python
### graficos de seaborn por paneles

sns.pairplot(datos)
```

```python
sns.jointplot(datos, x='density', y='fat', kind="hex", color="#4CB391")
```
